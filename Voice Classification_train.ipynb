```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('voice.csv')
```

```python
data.head()
```

```python
def convert_label(label):
    if label=="male":
        return 0
    else:
        return 1
```

```python
new_map=data['label'].map(convert_label)
data['label']=new_map
```

```python
data.head()
```

```python
X=data.iloc[:,0:20].values
Y=data.iloc[:,20:21].values

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
parameters={'C':[0.0001,0.001,0.01,0.1,1,10,100,1000]}
lr=LogisticRegression()
clf=GridSearchCV(lr,parameters,cv=5)
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
import pickle
pickle.dump(clf,open('voice_model.pickle','wb'))
```

```python
model=pickle.load(open('voice_model.pickle','rb'))
```

```python
model.score(X_train,y_train)
```
